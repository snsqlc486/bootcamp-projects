# 01. 데이터 수집 및 전처리

## 이 노트북의 목적
TMDB와 IMDb 두 사이트에서 영화/드라마 데이터를 수집하고, 분석에 쓸 수 있도록 전처리합니다.

## 전체 흐름

```
[1단계] TMDB 데이터 수집
  ├── 영화 ID 수집 (특정 기간 내)
  ├── 영화 상세 정보 수집 (장르, OTT 제공사, 투표 수 등)
  ├── 드라마 ID 수집
  └── 드라마 시리즈 + 시즌 상세 정보 수집

[2단계] IMDb 데이터 수집
  ├── 리뷰 텍스트 수집 (GraphQL API, 비동기 처리)
  └── 평점/투표수 수집 (HTML 스크래핑)

[3단계] 전처리
  ├── 데이터 정규화 (장르/언어/OTT 테이블 분리)
  ├── 리뷰 노이즈 제거 (HTML 태그, 짧은 리뷰 등)
  └── 흥행 점수 계산 (평점 + 감성 점수 → hit_score)
```

> ⚠️ 실제 전체 데이터 수집 코드는 파일 크기가 큰 관계로, 이 노트북은 **시험용(소규모)** 코드로만 구성됩니다.

# 기초 데이터 수집

영화와 드라마의 메타데이터(제목, 장르, 개봉일, OTT 제공사 등)를 TMDB API로 수집하고,  
사용자 리뷰와 평점을 IMDb에서 수집합니다.

## TMDB 데이터 수집

TMDB Discover API를 사용합니다.  
API 제한(최대 500페이지)으로 인해 기간을 N개월씩 나누어 수집하고, 멀티스레드로 속도를 높입니다.

````python
import file_collector.tmdb_collector as tdc  # TMDB API 수집 모듈 (영화·TV 메타데이터)
import time
import pandas as pd
from concurrent.futures import ThreadPoolExecutor, as_completed  # 여러 기간을 병렬로 처리
from tqdm import tqdm  # 루프 진행 상황을 시각적으로 표시
import warnings

warnings.filterwarnings("ignore")  # 불필요한 경고 메시지 숨김
````

````python
start_date = '2025-11-01'  # 수집 시작 날짜
end_date = '2025-11-10'    # 수집 종료 날짜

OUTPUT_DIR = "files"  # 수집한 데이터를 저장할 최상위 디렉터리
````

### 영화 수집

1. `collect_movie_ids()` 로 기간 내 영화 ID 목록 수집  
2. `fetch_movie_details()` 로 각 ID의 상세 정보를 멀티스레드(30 workers)로 병렬 수집  
3. 결과를 CSV로 저장

````python
if __name__ == "__main__":
    total_start = time.time()

    # 기간을 3개월 단위로 분할
    periods = tdc.generate_date_periods(start_date, end_date, months=3)

    print(f"\n전체 기간: {start_date} ~ {end_date}")
    print(f"분할 기간: {len(periods)}개")
    print("="*90)

    # ID 데이터를 집합으로 변환하여 중복 제거
    all_ids = set()

    for idx, (period_start, period_end) in enumerate(periods, 1):
        print(f"\n[{idx}/{len(periods)}] 기간: {period_start} ~ {period_end}")

        # 기간내 영화 ID 수집
        current_ids_set = tdc.collect_movie_ids(period_start, period_end)
        # ID를 집합에 추가
        all_ids.update(current_ids_set)

        print(f"해당 기간 내 수집된 ID: {len(current_ids_set):,}개 (누적 ID: {len(all_ids):,}개)")

    # 집합을 리스트로 변환
    id_list = list(all_ids)

    print(f"\n{'='*90}")
    print(f"최종 ID 개수: {len(id_list):,}개")

    print(f"\n{'='*90}")
    print("상세 정보 멀티스레드 수집 시작...\n")

    results = []
    # 3. 상세 정보 수집 단계: ID 리스트를 사용
    with ThreadPoolExecutor(max_workers=30) as executor:
        # 딕셔너리 대신 ID 자체를 future에 매핑
        futures = {executor.submit(tdc.fetch_movie_details, movie_ids): movie_ids for movie_ids in id_list}

        for future in tqdm(as_completed(futures), total=len(futures), desc="영화 상세 수집"):
            detail = future.result()
            if detail:
                results.append(detail)

    print(f"\n상세 정보 수집 완료: {len(results):,}개")

    # 저장
    df = pd.DataFrame(results)
    # 컬럼 이름이 'first_air_date'라고 가정
    df = df.sort_values(["release_date", "popularity"], ascending=[True, False])
    df.to_csv(f"{OUTPUT_DIR}/movies_test.csv", index=False, encoding="utf-8-sig")

    elapsed = time.time() - total_start

    print("\n" + "="*90)
    print("========================== DONE ==========================")
    print(f"총 데이터: {len(df):,}개")
    print(f"소요시간: {elapsed/60:.2f}분 ({elapsed:.2f}초)")
    print(f"저장 파일: movies.csv")
    print("="*90)
````

### 드라마 수집

영화와 달리 드라마는 **시리즈 + 시즌** 두 레벨로 구성됩니다.  
`fetch_tv_series_and_seasons()` 를 사용하여 한 번에 시리즈 정보와 시즌 목록을 동시에 수집합니다.  
결과는 `tv_series_test.csv` (시리즈)와 `tv_seasons_test.csv` (시즌) 두 파일로 저장됩니다.

````python
if __name__ == "__main__":
    total_start = time.time()

    # 기간을 N개월 단위로 분할
    periods = tdc.generate_date_periods(start_date, end_date, months=12)

    print(f"\n전체 기간: {start_date} ~ {end_date}")
    print(f"분할 기간: {len(periods)}개")
    print("=" * 90)

    # ID 데이터를 집합으로 변환하여 중복 제거
    all_ids = set()

    for idx, (period_start, period_end) in enumerate(periods, 1):
        print(f"\n[{idx}/{len(periods)}] 기간: {period_start} ~ {period_end}")

        # 기간내 TV 시리즈 ID 수집
        current_ids_set = tdc.collect_tv_ids(period_start, period_end)
        all_ids.update(current_ids_set)

        print(f"해당 기간 내 수집된 ID: {len(current_ids_set):,}개 (누적 ID: {len(all_ids):,}개)")

    # 집합을 리스트로 변환
    id_list = sorted(list(all_ids))

    print(f"\n{'=' * 90}")
    print(f"최종 ID 개수: {len(id_list):,}개")

    # ============================================================
    # 2단계: TV 시리즈 + 시즌 상세 정보 수집 (멀티스레드)
    # ============================================================
    print(f"\n{'=' * 90}")
    print("상세 정보 멀티스레드 수집 시작...\n")

    series_results = []
    seasons_results = []

    with ThreadPoolExecutor(max_workers=20) as executor:
        # 시리즈 + 시즌 동시 수집
        futures = {
            executor.submit(tdc.fetch_tv_series_and_seasons, tv_id): tv_id
            for tv_id in id_list
        }

        for future in tqdm(as_completed(futures), total=len(futures), desc="TV 시리즈 + 시즌 수집"):
            try:
                series_data, seasons_list = future.result()

                if series_data:
                    series_results.append(series_data)

                if seasons_list:
                    seasons_results.extend(seasons_list)

            except Exception as e:
                tv_id = futures[future]
                print(f"오류 (TV ID: {tv_id}): {e}")

    print(f"\n시리즈 정보 수집 완료: {len(series_results):,}개")
    print(f"시즌 정보 수집 완료: {len(seasons_results):,}개")

    # ============================================================
    # 3단계: 저장
    # ============================================================

    # 시리즈 데이터 저장
    df_series = pd.DataFrame(series_results)
    df_series = df_series.sort_values(["first_air_date", "popularity"], ascending=[True, False])

    # 시즌 데이터 저장
    df_seasons = pd.DataFrame(seasons_results)
    df_seasons = df_seasons.sort_values(["series_id", "season_number"], ascending=[True, True])

    # CSV 저장
    df_series.to_csv(f"{OUTPUT_DIR}/tv_series_test.csv", index=False, encoding="utf-8-sig")
    df_seasons.to_csv(f"{OUTPUT_DIR}/tv_seasons_test.csv", index=False, encoding="utf-8-sig")

    elapsed = time.time() - total_start

    print("\n" + "=" * 90)
    print("========================== DONE ==========================")
    print(f"시리즈 데이터: {len(df_series):,}개")
    print(f"시즌 데이터: {len(df_seasons):,}개")
    print(f"소요시간: {elapsed / 60:.2f}분 ({elapsed:.2f}초)")
    print(f"저장 파일: tv_series_test.csv, tv_seasons_test.csv")
    print("=" * 90)
````

## 흥행 점수 계산

**hit_score** 는 작품의 종합 흥행 지표입니다.

```
hit_score = ((scaled_rating + sentiment_score) / 2) × scaled_num_votes × 100
```

- `scaled_rating`: IMDb 평점을 0~1로 정규화
- `sentiment_score`: 리뷰 긍정 비율 (BERT 감성 분석)
- `scaled_num_votes`: 투표 수를 로그 변환 후 0~1로 정규화 (인기도 반영)

처리 순서:
1. `SentimentAnalyzer` 로 리뷰 감성 분석 실행 (청크 단위, Resume 지원)
2. `compute_weighted_sentiment_score()` 로 작품별 가중 감성 점수 집계
3. `calculate_ratings()` 로 IMDb 평점 정규화
4. `calculate_score()` 로 최종 hit_score 계산

````python
import preprocessing.scoring as sc   # 흥행 점수(hit_score) 계산 모듈
import preprocessing.sentiment as sp  # 감성 분석 점수 계산 모듈
````

````python
# 영화·드라마 메인 데이터에서 평점 관련 컬럼만 선택
movie_main = pd.read_parquet(f"{OUTPUT_DIR}/movie/00_movie_main.parquet")
tv_main = pd.read_parquet(f"{OUTPUT_DIR}/drama/00_tv_main.parquet")
main = pd.concat([
    movie_main[['imdb_id', 'tmdb_rating', 'tmdb_rating_count', 'imdb_rating', 'imdb_rating_count']],
    tv_main[['imdb_id', 'tmdb_rating', 'tmdb_rating_count', 'imdb_rating', 'imdb_rating_count']]
])  # 영화·드라마 평점 데이터를 수직으로 합침
review = pd.read_parquet(f"{OUTPUT_DIR}/reviews/reviews_test.parquet")  # 정제된 리뷰 로드
````

````python
analyzer = sp.SentimentAnalyzer()  # 감성 분석기 초기화 (긍정/부정 점수 산출)
analyzer.analyze(review, f"{OUTPUT_DIR}/reviews", text_column="review_text")  # 리뷰별 감성 점수 계산 및 저장
````

````python
# 감성 분석 결과(chunk)와 원본 리뷰의 helpful_ratio를 review_id로 병합
review = pd.read_parquet(f"{OUTPUT_DIR}/reviews/review_chunk_00000.parquet").merge(pd.read_parquet(f"{OUTPUT_DIR}/reviews/reviews_test.parquet")[['helpful_ratio', 'review_id']], on='review_id', how='left')
````

````python
sentiment_score = sc.compute_weighted_sentiment_score(review)  # 유용도(helpful_ratio) 가중 감성 점수
main_score = sc.calculate_ratings(main)                        # TMDB·IMDb 평점 기반 종합 점수
````

````python
# 평점 점수 + 감성 점수를 결합하여 최종 흥행 점수(hit_score) 계산·저장
sc.calculate_score(main_score.merge(sentiment_score, on='imdb_id', how='inner'))
````
